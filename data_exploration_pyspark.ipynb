```python
import pyspark.pandas as ps
import os
```

```python
base_df = ps.DataFrame()
folder_path = './data'
for filename in os.listdir(folder_path):
    if filename.endswith(".csv"):
        # Prints only text file present in My Folder
        print('report filename:', filename)
        file_path = os.path.join(folder_path, filename)
        try:
            # read day report df 
            day_report_df = ps.read_csv(file_path)
        except Exception as e:
            # read day report df using latin encoding
            day_report_df = ps.read_csv(file_path, encoding='latin1')
        finally:
            # rename columns, remove blank spaces and replace them by '_'
            columns_dict = {}
            for column_report in day_report_df.columns:
                columns_dict[column_report] = column_report.strip().replace(' ', '_').lower()
            day_report_df.rename(columns=columns_dict, inplace=True)
            # concatenate report
            base_df = ps.concat([base_df, day_report_df], ignore_index=True)
```
